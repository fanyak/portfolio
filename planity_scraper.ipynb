```python
import pandas as pd
import json
```

```python
a = !shot-scraper javascript https://www.planity.com/barbier/paris-75 '(() => { function tableToObjects() {const rows = Array.from(document.querySelectorAll(".business_item_search-module_business-YTATk")); const obj = {}; for (let i = 0; i < rows.length; i++) { const row = rows[i]; const title = row.querySelector(".business_item_search-module_title-ryICj").innerHTML; const address = Array.from(row.querySelectorAll("#address-value span")).map(span => span.innerText.trim()).join(" ").trim(); obj[title] = address; } return obj;} return tableToObjects();})'
```

```python
data = json.loads(a.s)
```

```python
df = pd.DataFrame.from_dict(data, orient='index', columns=['Full Address'])
df.loc[:, "Arrondissements"] = df["Full Address"].str.split(",", n=2, expand=True)[1]
df["Arrondissements"] = df["Arrondissements"].str.extract(r'([\d]{1,5})')
```

```python
df.to_excel("planity.xlsx")
```
